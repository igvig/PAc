```python
import tensorflow as tf
import pandas as pd
import numpy as np
import os

approximator = PAc_Approximator(
    func = lambda x: tf.nn.gelu(x).numpy(),
    func_low = lambda x: 0.,
    func_high = lambda x: x,
)

for f in os.listdir('.'):
    if f.endswith('.csv') and f.startswith('gelu'):
        # Get the table size from the filename
        table_size = int(f.split('_')[1].split('.')[0])
        df = pd.read_csv(f)
        # scan through each row and get the compount, auc, and max errors for the xmin-xmax range of each row, and add these values to the dataframe
        for i, row in df.iterrows():
            xmin, xmax = row['x_min'], row['x_max']
            compound_error, auc_error, max_error = approximator.approx_error(xmin, xmax, table_size)
            print(f"t-size: {table_size}, xmin: {xmin}, xmax: {xmax}, original compound error: {df.at[i, 'error']}, compound_error: {compound_error}, auc_error: {auc_error}, max_error: {max_error}")
            df.at[i, 'error_auc'] = auc_error
            df.at[i, 'error_max'] = max_error
        df.to_csv(f, index=False)
```
